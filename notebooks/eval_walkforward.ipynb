# Walk-forward Evaluation (v1)

Loads walk-forward results and visualizes:
- Sharpe by year
- Total return by year
- Max drawdown by year

Input:
- data/results/walkforward_results.csv


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
res = pd.read_csv("../data/results/walkforward_results.csv")
res
```

```python
plt.figure(figsize=(10, 4))
plt.plot(res["test_year"], res["sharpe"], marker="o")
plt.title("Walk-forward Sharpe by Test Year")
plt.xlabel("Test Year")
plt.ylabel("Sharpe (approx)")
plt.axhline(0, linewidth=1)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(res["test_year"], res["total_return"], marker="o")
plt.title("Walk-forward Total Return by Test Year")
plt.xlabel("Test Year")
plt.ylabel("Total Return")
plt.axhline(0, linewidth=1)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(res["test_year"], res["max_drawdown"], marker="o")
plt.title("Walk-forward Max Drawdown by Test Year")
plt.xlabel("Test Year")
plt.ylabel("Max Drawdown")
plt.axhline(0, linewidth=1)
plt.show()
```

```python
summary = res[["sharpe", "total_return", "max_drawdown"]].describe()
summary
```
